### Import of Libaries

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Implemantation of the Linear Regression Calculation

```python
class LinearRegression:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def correlation(self):
        return (np.cov(self.x, self.y, bias=True)[0][1]) / np.sqrt((np.var(self.x) * np.var(self.y)))
    
    def inclination(self):
        return self.correlation() * (np.std(self.y) / np.std(self.x))
    
    def interceptation(self):
        return np.mean(self.y) - self.inclination() * np.mean(self.x)
    
    def prevision(self, value):
        return self.interceptation() + (self.inclination() * value)
    
    def showLinearRegression(self, title_graph='Linear Regression', name_axis_x='Axis X', name_axis_y='Axis Y'):
        plt.xlabel(name_axis_x)
        plt.ylabel(name_axis_y)
        plt.title(title_graph)

        plt.scatter(self.x, self.y)
        plt.plot([np.min(self.x), np.max(self.x)], [self.prevision(np.min(self.x)), self.prevision(np.max(self.x))], c='orange')
        plt.show()
```

### Test

```python
age = [18, 23, 25, 33, 34, 43, 48, 51, 58, 63, 67]
value = [871, 1100, 1393, 1654, 1915, 2100, 2356, 2698, 2959, 3000, 3100]

rel = LinearRegression(age, value)
prev = rel.prevision(105)

prev
```
